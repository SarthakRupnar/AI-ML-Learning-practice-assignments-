```python
import json
```

```python
def load_file(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data = load_file("store_data.json")
print(data)
```

```python
def clean_data(data):
    text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data = []
    unique_data = set()
    for user in data:
        #clean ratings
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        #handling missing vals
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        if(user["name"].strip() in unique_data):
            continue
        unique_data.add(user["name"])
        cleaned_data.append(user)
    return cleaned_data
```

```python
data = clean_data(data)
```

```python
print(data)
```

```python
def get_avg_rat(data):
    total_rat = 0
    for user in data:
        total_rat = total_rat + float(user["rating"])
    print(f"average rating is:{total_rat/len(data)}")
```

```python
get_avg_rat(data)
```

```python
def get_poor_rat_per(data):
    poor_rat = 0
    for user in data:
        if(float(user["rating"]) < 3):
             poor_rat += 1
    print(f"poor rating is:{poor_rat/len(data) * 100}")
```

```python
get_poor_rat_per(data)
```
