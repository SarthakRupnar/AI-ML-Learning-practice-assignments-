```python
import matplotlib.pyplot as plt
```

```python
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]

oscar_revenue = [1005, 170, 427, 133, 232] 

non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]

non_oscar_revenue = [378, 2788, 829, 185, 275]
```

```python
plt.plot(years,oscar_revenue,"o--r",label ="oscar movies")
plt.plot(years,non_oscar_revenue,"H--g",label ="non oscar movies")
plt.title("Oscar Movies vs Non-Oscar Movies")
plt.xlabel("Movies")
plt.ylabel("Revenue in M$")

plt.legend()
plt.show()
```

```python
plt.plot(years,oscar_revenue,label ="oscar movies")
plt.plot(years,non_oscar_revenue ,label ="non oscar movies")
plt.title("Oscar Movies vs Non-Oscar Movies")
plt.xlabel("Movies")
plt.ylabel("Revenue in M$")

plt.legend()

print(plt.style.available)
plt.style.use('seaborn-v0_8-notebook')
plt.show()
```
