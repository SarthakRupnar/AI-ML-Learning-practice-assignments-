```python
import requests

for i in range(1,11):
    URL = f"https://quotes.toscrape.com/page/{i}/"
    res = requests.get(URL)
    res
    with open("scraped_data/activity1_data.html","a", encoding="utf-8") as f:
        f.write(res.text)
    print(res.text)
```

```python
from bs4 import BeautifulSoup

with open("scraped_data/activity1_data.html","r", encoding="utf-8") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content,"lxml")
# print(html_content)
```

```python
all_quotes = soup.find_all("div", class_="quote")

for quote_div in all_quotes:

    tags = quote_div.find_all("a", class_="tag")

    for tag in tags:

        if tag.text.strip() == "life":

            quote = quote_div.find("span", class_="text")
            author = quote_div.find("small", class_="author")

            print("Quote:", quote.text)
            print("Author:", author.text)
```
