```python
df2 = df.copy()
df2 = df2.fillna(0)
df2["age"] = df2["age"].astype("int64")
```

```python
df2.dtypes
```

```python
df2
```

```python
# feature engineering

df2["tax"] = df2["income"].apply(lambda x:"20%" if x>=50000 else "10%")
df2
```

```python
df2["annual income"] = df2["income"]*12
df2
```

```python
gender_map = {
    "Male":"M",
    "Female":"F",
    "Unknown":"U"
}
df2["gender"] = df2["gender"].map(gender_map)
df2
```

```python
df2 = df2.assign(new_income = df2["income"]*1.1)
df2
```

```python
df2["country"] = df2["country"].replace("USA","America")
df2
```

```python
df2.columns = ["Id","Name","Age","Country","Gender","Income","Tax","Annual Income","New Income"]
df2
```

```python
df2 = df2.rename(columns={"Income":"Salary"})
df2
```

```python
df2["Salary"].sort_values()
```

```python
df2.sort_values("Salary")
```

```python
df2["Income Ranking"] = df2["Salary"].rank(ascending = False , method = "max") 
df2
```
